This file shows how to:
- open and plot the contents of processed .npz files (full domain and basin only velocities)
- calculate timeseries of basin-integrated kinetic energy

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load grid data

grid_data = np.load('../data/processed_data/grid_info.npz')
time = grid_data['time'] # time of each save time step
x_basin0 = grid_data['x_basin0'] # grid cell of the left extent of the basin 
x_basin1 = grid_data['x_basin1'] # grid cell of the right extent of the basin
y_inlet = grid_data['y_inlet'] # grid cell of the north extent of the basin
hFacc = grid_data['hFacc'] # partial cell fill (cell-centered)
drf = grid_data['drf'] # vertical spacing
hFacw = grid_data['hFacw'] # partial cell fill (west edge)
hFacs = grid_data['hFacs'] # partial cell fill (south edge)
z = grid_data['z'] # array of z values
y = grid_data['y'] # array of y values
x = grid_data['x'] # array of x values
depth = grid_data['depth'] # 2D array of depth
```

```python
# load u-velocity for the entire domain (at the last time step) and plot the surface values

full_domain_data = np.load('../data/processed_data/last_time_step.npz')
uvel_full = full_domain_data['uvel'] # shape is [nz, ny, nx]

fig, axs = plt.subplots() 
im1 = axs.pcolor(x/1e3,y/1e3,uvel_full[0,:,1:], shading='auto',cmap='bwr',vmin=-0.02,vmax=0.02)
axs.set(xlabel='$x$ (km)', ylabel='$y$ (km)',
       title='Full domain $u$ (m/s) at last time step')
fig.colorbar(im1)
fig.set_size_inches(12, 3)
```

```python
# load velocity for the basin for all time steps and plot the surface values of u-velocity (at the last time step)
#      with added bathymetry contours every 500 m

basin_data = np.load('../data/processed_data/basin_data.npz')
uvel_basin = basin_data['uvel_basin'] # shape is [nt, nz, ny, nx]
vvel_basin = basin_data['vvel_basin'] # shape is [nt, nz, ny, nx]
wvel_basin = basin_data['wvel_basin'] # shape is [nt, nz, ny, nx]

fig, axs = plt.subplots()
im1 = axs.pcolor(x[x_basin0:x_basin1]/1e3,y[:y_inlet]/1e3,
                 uvel_basin[-1,0,:,:], shading='auto',cmap='bwr',vmin=-0.02,vmax=0.02)
axs.contour(x[x_basin0:x_basin1]/1e3,y[:y_inlet]/1e3,depth[:y_inlet,x_basin0:x_basin1],
           levels=[500,1e3,1.5e3,2e3,2.5e3,3e3,3.5e3],colors='k')
axs.set(xlabel='$x$ (km)', ylabel='$y$ (km)',
       title='Basin surface $u$ (m/s) at last time step with bathymetry contours')
fig.colorbar(im1)
fig.set_size_inches(8, 6)
```

Now, we will show how to compute basin-integrated kinetic energy at each timestep and plot its timeseries. 

Because we compute baroclinic kinetic energy, so we will subtract the barotropic (depth-averaged) velocity component.

```python
# Compute barotropic velocities (which are defined to be depth-averaged quantities)

def u_bt(u,hf,drf,depth):
    Uw = u*hf*drf[:,None,None]
    uc = np.nansum(Uw,axis=0)/depth
    return uc

def v_bt(v,hf,drf,depth):
    Vw = v*hf*drf[:,None,None]
    vc = np.nansum(Vw,axis=0)/depth
    return vc
```

```python
dx = np.mean(np.diff(x[x_basin0:x_basin1]))
dy = np.mean(np.diff(y[:y_inlet]))
vol = dx*dy*hFacc[:, :y_inlet, x_basin0:x_basin1]*drf[:,None,None]
vol = np.nansum(vol)
rho_0 = 1000.
    
# loop over all timesteps and compute
ke_bc_basin = 0*time
    
for i in range(1, len(time)):
    u = uvel_basin[i,:,:,:]
    v = vvel_basin[i,:,:,:]
    w = wvel_basin[i,:,:,:]
            
    # barotropic u, v velocities    
    ubt = u_bt(u,hFacw[:, :y_inlet, x_basin0:x_basin1],drf,depth[:y_inlet, x_basin0:x_basin1])
    vbt = v_bt(v,hFacs[:, :y_inlet, x_basin0:x_basin1],drf,depth[:y_inlet, x_basin0:x_basin1])
      
    # baroclinic u, v velocities: subtracted barotropic component from full field    
    upc = u - ubt[None,:,:]
    vpc = v - vbt[None,:,:]
        
    # calculate and integrate baroclinic kinetic energy and save per timestep    
    KEpc = 0.5*rho_0*(upc**2 + vpc**2 + w**2)    
    Ebc = np.nansum(KEpc*hFacc[:, :y_inlet, x_basin0:x_basin1]*drf[:,None,None],axis=0)
    ke_bc_basin[i] = dx*dy*np.nansum(Ebc)
```

```python
# Plot timeseries of baroclinic kinetic energy

omega = 1.4136e-4

fig, ax = plt.subplots() 
ax.plot(time/(2*np.pi/omega),ke_bc_basin/vol)
ax.set(xlabel='Forcing period time $t_f=2\pi/\omega$', 
       ylabel='Basin-average baroclinic KE $(J/m^3)$',
       title='Timeseries for sample simulation')
```
